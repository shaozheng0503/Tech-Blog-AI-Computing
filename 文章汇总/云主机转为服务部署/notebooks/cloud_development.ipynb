```python
# ğŸš€ å¯åŠ¨APIæœåŠ¡å™¨å¹¶æµ‹è¯•
import subprocess
import time
import requests
import threading

def start_api_server():
    """å¯åŠ¨APIæœåŠ¡å™¨"""
    print("ğŸš€ å¯åŠ¨äº‘ä¸»æœºAPIæœåŠ¡å™¨...")
    
    try:
        from src.app import app
        import uvicorn
        
        print(f"ğŸ“ APIæœåŠ¡åœ°å€:")
        print(f"   ğŸ  é¦–é¡µ: https://{CLOUD_HOST}")
        print(f"   ğŸ“– APIæ–‡æ¡£: https://{CLOUD_HOST}/docs")
        print(f"   ğŸ“š ReDoc: https://{CLOUD_HOST}/redoc")
        print(f"   ğŸ“Š ç»Ÿè®¡: https://{CLOUD_HOST}/stats")
        
        # åœ¨åå°å¯åŠ¨æœåŠ¡å™¨
        def run_server():
            uvicorn.run(app, host="0.0.0.0", port=API_PORT, log_level="info")
        
        server_thread = threading.Thread(target=run_server, daemon=True)
        server_thread.start()
        
        # ç­‰å¾…æœåŠ¡å™¨å¯åŠ¨
        time.sleep(3)
        print("âœ… APIæœåŠ¡å™¨å·²å¯åŠ¨ï¼")
        
        return True
        
    except Exception as e:
        print(f"âŒ å¯åŠ¨å¤±è´¥: {e}")
        return False

# å¯åŠ¨æœåŠ¡å™¨
if start_api_server():
    # æµ‹è¯•APIè¿æ¥
    try:
        response = requests.get(f"https://{CLOUD_HOST}/health", timeout=5)
        if response.status_code == 200:
            print("âœ… APIè¿æ¥æµ‹è¯•æˆåŠŸ")
            print(f"ğŸ“Š å¥åº·çŠ¶æ€: {response.json()}")
        else:
            print(f"âš ï¸ APIè¿æ¥å¼‚å¸¸: {response.status_code}")
    except Exception as e:
        print(f"âš ï¸ è¿æ¥æµ‹è¯•å¤±è´¥: {e}")
        print("è¯·ç¡®ä¿äº‘ä¸»æœºé˜²ç«å¢™å…è®¸è®¿é—®")
```

```python
# ğŸ§ª APIåŠŸèƒ½æµ‹è¯•å’Œæ¼”ç¤º
import json
from IPython.display import display, HTML, JSON

def test_api_endpoints():
    """æµ‹è¯•æ‰€æœ‰APIç«¯ç‚¹"""
    base_url = f"https://{CLOUD_HOST}"
    
    print(f"ğŸ§ª APIåŠŸèƒ½æµ‹è¯• - {base_url}")
    print("=" * 60)
    
    tests = [
        ("å¥åº·æ£€æŸ¥", "GET", "/health"),
        ("APIç»Ÿè®¡", "GET", "/stats"),
        ("ç”¨æˆ·åˆ—è¡¨", "GET", "/users?limit=5"),
        ("æœç´¢ç”¨æˆ·", "GET", "/users/search/äº‘"),
        ("è·å–ç”¨æˆ·è¯¦æƒ…", "GET", "/users/1"),
        ("å¹´é¾„èŒƒå›´æŸ¥è¯¢", "GET", "/users/age-range/25/35")
    ]
    
    results = []
    
    for test_name, method, endpoint in tests:
        try:
            url = f"{base_url}{endpoint}"
            print(f"\nğŸ” {test_name}")
            print(f"   URL: {url}")
            
            start_time = time.time()
            response = requests.get(url, timeout=10)
            response_time = (time.time() - start_time) * 1000
            
            if response.status_code == 200:
                data = response.json()
                print(f"   âœ… æˆåŠŸ ({response.status_code}) - {response_time:.1f}ms")
                
                # æ˜¾ç¤ºå“åº”æ‘˜è¦
                if 'data' in data:
                    if isinstance(data['data'], dict):
                        if 'users' in data['data']:
                            print(f"   ğŸ“Š ç”¨æˆ·æ•°é‡: {len(data['data']['users'])}")
                        elif 'count' in data['data']:
                            print(f"   ğŸ“Š åŒ¹é…æ•°é‡: {data['data']['count']}")
                        else:
                            print(f"   ğŸ“‹ æ•°æ®å­—æ®µ: {list(data['data'].keys())}")
                
                results.append({
                    'test': test_name,
                    'success': True,
                    'response_time': response_time,
                    'data': data
                })
            else:
                print(f"   âŒ å¤±è´¥ ({response.status_code})")
                results.append({
                    'test': test_name,
                    'success': False,
                    'status_code': response.status_code
                })
                
        except Exception as e:
            print(f"   âŒ å¼‚å¸¸: {str(e)}")
            results.append({
                'test': test_name,
                'success': False,
                'error': str(e)
            })
    
    # æµ‹è¯•ç»“æœæ±‡æ€»
    successful = [r for r in results if r.get('success', False)]
    print(f"\nğŸ“Š æµ‹è¯•ç»“æœæ±‡æ€»:")
    print(f"   æ€»æµ‹è¯•æ•°: {len(results)}")
    print(f"   æˆåŠŸ: {len(successful)}")
    print(f"   å¤±è´¥: {len(results) - len(successful)}")
    print(f"   æˆåŠŸç‡: {len(successful)/len(results)*100:.1f}%")
    
    if successful:
        avg_time = sum(r.get('response_time', 0) for r in successful) / len(successful)
        print(f"   å¹³å‡å“åº”æ—¶é—´: {avg_time:.1f}ms")
    
    return results

# è¿è¡ŒAPIæµ‹è¯•
test_results = test_api_endpoints()
```
